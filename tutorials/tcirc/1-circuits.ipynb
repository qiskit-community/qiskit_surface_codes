```python
from qtcodes import TopologicalRegister, TopologicalCircuit, REPETITION, XXZZ, XZZX
from qiskit import execute, Aer, QuantumCircuit, QuantumRegister
```

```python
treg = TopologicalRegister(3, ctype=REPETITION)
tcirc = TopologicalCircuit(treg)
tcirc.x(treg[0])
tcirc.stabilize(0)
tcirc.measure_z(0)
tcirc.draw(output='mpl', fold=500)
```

## Simulate

```python
from qiskit import execute, Aer
def sim(circ):
    return execute(circ, 
                  Aer.get_backend("aer_simulator"), 
                  shots=10,
                 ).result().get_counts()
```

```python
print(sim(tcirc.circ))
```

## CNOT

```python
treg = TopologicalRegister(2, ctype=REPETITION)
tcirc = TopologicalCircuit(treg)

# prep state
tcirc.reset_z(0)
tcirc.reset_z(1)
tcirc.x(0)
tcirc.x(1)

# cnot
tcirc.cx(0,1)

# meas
tcirc.measure_z(0)
tcirc.measure_z(1)

results = sim(tcirc.circ)
results = set([result[:3] for result in results.keys()])
print("q1 q0")
print(results)
```

```python
tcirc.draw(output='mpl', fold=500)
```
